```python
import os
import numpy as np
import open3d as o3d
from PIL import Image
import matplotlib.pyplot as plt

# --- Configuration ---
# This path should be correct if you downloaded the 78 MB file
kitti_base_dir = '../data/2011_09_26_drive_0013_sync/'

image_dir = os.path.join(kitti_base_dir, 'image_02/data/')
lidar_dir = os.path.join(kitti_base_dir, 'velodyne_points/data/')

# --- 1. Load and Display Camera Image ---
sample_image_file = os.path.join(image_dir, '0000000000.png')
image = Image.open(sample_image_file)

print(f"Displaying image: {sample_image_file}")
plt.figure(figsize=(12, 6))
plt.imshow(image)
plt.title("Sample Camera Image")
plt.axis('off')
plt.show()

# --- 2. Load and Visualize LiDAR Point Cloud ---
sample_lidar_file = os.path.join(lidar_dir, '0000000000.bin')
point_cloud = np.fromfile(sample_lidar_file, dtype=np.float32).reshape(-1, 4)
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(point_cloud[:, :3])

print(f"\nVisualizing LiDAR scan: {sample_lidar_file}")
print("Close the pop-up window to continue.")

o3d.visualization.draw_geometries([pcd],
                                  window_name="LiDAR Point Cloud",
                                  point_show_normal=False)
```
